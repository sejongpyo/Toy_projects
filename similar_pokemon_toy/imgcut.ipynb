```python
import os, glob, random
import matplotlib.pyplot as plt
import cv2

def crop_face(path) :
    cascade_file = "haarcascade_frontalface_alt.xml" # 정면 얼굴 검출
    cascade = cv2.CascadeClassifier(cascade_file)
    img = cv2.imread(path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_list = cascade.detectMultiScale(img_gray, minSize=(50,50)) # miniSize - 얼굴 인식 영역의 최소 크기 지정\
    for (x,y,w,h) in face_list:
        red = (0, 0, 255) 
        dst = img.copy()####이미지복사해서
        dst = img[y:y+h,x:x+w] ###자르기
    return dst
```

```python
files = glob.glob("/Users/flora/Desktop/project/aiproject/chicolita/beak" + "/*.png")  ### 폴더지정 : 폴더이름이 영어야 한다 na solar yeri
num = 0
# bek irene jisoo sej sonna
for f in files:
    if num >= 300: break  #개수지정
    try :
        res = crop_face(f)
        filename = str(num) + ".png"
        cv2.imwrite(filename,res)
    except:
        pass    
    num += 1
```
